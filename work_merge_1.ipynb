```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.next = None

class SLinkList(object):
    def __init__(self):
        self.head = None
        self.length = 0

    def is_empty(self):
        return self.head is None

    def add(self, p):
        if self.is_empty():
            self.head = p
        else:
            p.next = self.head
            self.head = p
        self.length += 1

    def appendnode(self, p):
        if self.is_empty():
            self.add(p)
        else:
            q = self.head
            while q.next is not None:
                q = q.next
            q.next = p
            self.length += 1

    def travel(self):
        q = self.head
        if self.length == 0:
            print("目前链表没有数据！")
        else:
            print("目前链表数据有：", end=" ")
            for _ in range(self.length):
                print(f"{q.data}->", end=" ")
                q = q.next
            print("\n")

def link_merge_sort(l1, l2):
    merged = SLinkList()
    while not l1.is_empty() and not l2.is_empty():
        if l1.head.data < l2.head.data:
            current = l1.head
            l1.head = current.next
            current.next = None
            merged.appendnode(current)
            l1.length -= 1
        else:
            current = l2.head
            l2.head = current.next
            current.next = None
            merged.appendnode(current)
            l2.length -= 1

    while not l1.is_empty():
        current = l1.head
        l1.head = current.next
        current.next = None
        merged.appendnode(current)
        l1.length -= 1

    while not l2.is_empty():
        current = l2.head
        l2.head = current.next
        current.next = None
        merged.appendnode(current)
        l2.length -= 1

    return merged
```

```python
# 创建链表1：1 -> 3 -> 5
link1 = SLinkList()
for data in [5, 3, 1]:
    link1.add(Node(data))

# 创建链表2：2 -> 4 -> 6
link2 = SLinkList()
for data in [6, 4, 2]:
    link2.add(Node(data))

# 合并链表
merged = link_merge_sort(link1, link2)
merged.travel()  # 输出：1->2->3->4->5->6
```

```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.next = None

class SLinkList(object):
    def __init__(self):
        self.head = None
        self.length = 0

    def is_empty(self):
        return self.head is None

    def add(self, p):
        if self.is_empty():
            self.head = p
        else:
            p.next = self.head
            self.head = p
        self.length += 1

    def appendnode(self, p):
        if self.is_empty():
            self.add(p)
        else:
            q = self.head
            while q.next is not None:
                q = q.next
            q.next = p
            self.length += 1

    def travel(self):
        q = self.head
        if self.length == 0:
            print("目前链表没有数据！")
        else:
            print("目前链表数据有：", end=" ")
            for _ in range(self.length):
                print(f"{q.data}->", end=" ")
                q = q.next
            print("\n")


def link_merge_sort_desc(l1, l2):
    merged = SLinkList()
    while not l1.is_empty() and not l2.is_empty():
        # 修改比较逻辑：选择较大的节点
        if l1.head.data > l2.head.data:  # <-- 将 < 改为 >
            current = l1.head
            l1.head = current.next
            current.next = None
            merged.appendnode(current)
            l1.length -= 1
        else:
            current = l2.head
            l2.head = current.next
            current.next = None
            merged.appendnode(current)
            l2.length -= 1

    # 处理剩余节点（逻辑保持不变）
    while not l1.is_empty():
        current = l1.head
        l1.head = current.next
        current.next = None
        merged.appendnode(current)
        l1.length -= 1

    while not l2.is_empty():
        current = l2.head
        l2.head = current.next
        current.next = None
        merged.appendnode(current)
        l2.length -= 1

    return merged
```

```python
# 创建降序链表 l1: 5→3→1（用头插法）
l1 = SLinkList()
for data in [1, 3, 5]:  # 注意添加顺序与链表顺序相反
    l1.add(Node(data))

# 创建降序链表 l2: 6→4→2
l2 = SLinkList()
for data in [2, 4, 6]:
    l2.add(Node(data))

# 合并链表
merged = link_merge_sort_desc(l1, l2)
merged.travel()  # 输出：6→5→4→3→2→1
```
